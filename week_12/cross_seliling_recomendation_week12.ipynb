```python

def plot(feature, data, colors='Blues', hue=None):
    """
    Plots a countplot with percentage labels.

    Parameters:
    - feature: Column name to plot
    - data: DataFrame containing the column
    - colors: Seaborn color palette
    - hue: Column to group by (optional)
    """
    ncount = len(data)
    
    if ncount == 0:
        print("No data available for plotting.")
        return

    fig, ax = plt.subplots(figsize=(8, 5))
    
    ax = sns.countplot(x=feature, data=train_data, palette=colors, hue=hue)

    ax.set_ylabel('Count')
    ax.set_title(f'Distribution of {feature}')

    # Add percentage labels
    for p in ax.patches:
        total = p.get_height()
        if total > 0:  # Avoid division by zero
            ax.annotate(f'{100.*total/ncount:.1f}%', 
                        (p.get_x() + p.get_width() / 2, total), 
                        ha='center', va='bottom', fontsize=11, color='black')

    plt.show()
```
